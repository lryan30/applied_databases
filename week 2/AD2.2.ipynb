##### **Relational Database:**

-   A relational database consists of a set of tables used for storing data. 
-   A table is a collection of related data. 
-   Each table has a unique name and may relate to one or more other tables in the database through common values. 


*A table in a database is a collection of rows and columns. Also known as entities or relations.*
-   A row contains data relating to a single item or record in a table. Also known as records or tuples. 
-   A column containing data representing a specific characteristic of the records in the table. Also known as fields or attributes. 

-   Each row in the spreadsheet contains data relating to a single item.(project). 
-   rows in a spreadsheet are the same as rows in a table in a relational db. 

-   Each column represents the specific characteristics of the project. 
-   columns in spreadsheet are the same as columns in a table in a relational db. 

Individual sheets of a spreadsheet correspond to tables of a database. Even though a spreadsheet is not a database, some of the concepts of a spreadsheet can be easily translated in relational database terms.

### **Database Schema:**

-   A database consists of schemas, tables, views and other objects. 
-   A database schema is usually a drawing that represents an idea or theory and makes it easier to understand. (diagram of car engine eg.). 
-   the logical configuration of all or part of a database. 
-   it defines how the data and relationships between the data is stored. 
-   most databases consist of more than one table and the schema shows how they are associated with one another. 

Two types of Schema: 

-   *physical schema:* 
defines how data is stored physicaly on a storage system in terms of files and indices. may take into consideration the datasize, does it need to be spread over multiple computers etc. 

-   *logical schema:*
defines the logical contraints that apply to the stored data, the tables in the database and the relationships between them.  (eg: customer name should be no more than 50 characters long or the price of a product can be stored as a double rather than an integer.)
-   The logical schema is designed before the database is created. 
-   No data is contained in the logical schema.

### Database Management System: 

-   software for creating and managing databases. 
-   interacts with the user, the database itself and other systems in order to store, retrieve and process data. (OS FOR DB) 
-   provides a centralised view of data that can be accessed by mutiple users, from multiple locations in a controlled manner. concurrency-2 or more users accessing data at same time. 
-   the DBMS can limit what data the end user sees, as well as how that end user can view the data, providing many views of a single database schema. 

example: store employee details
-   dbms can manage different types of users to access the table. 
-   hr can probably see all employee details
-   manaager might not need to know all the details 
-   dbms provides many different views on the same data. 



The DBMS provides data independence,(so that the underlying hardware or storage implementation changes the user is not affected.) freeing users and application programs from knowing where or how the data is stored. 
-   Any changes in how or where the data is stored is completely transparent due to the DBMS. 
-   CRUD [Create, Read, Update, Function] functions of a DBMS.
